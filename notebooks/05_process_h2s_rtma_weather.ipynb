```python
# Matches stronger H₂S events with RTMA weather variables
# and creates wind-speed and wind-direction summaries.

from pathlib import Path
import pandas as pd
import numpy as np

# Mount Google Drive in Colab
from google.colab import drive
drive.mount("/content/drive")

# Main project folders
DATA_RAW = Path("/content/drive/MyDrive/STAT596/Project596_datafiles")
DATA_PROCESSED = DATA_RAW / "processed_data"
FIGURES = DATA_RAW / "figures"
TABLES = DATA_RAW / "tables"
NOTEBOOKS = DATA_RAW / "notebooks"

# Input files
current_h2s_path = DATA_PROCESSED / "h2s_events_clean_2024_2025.csv"
rtma_matched_input_path = DATA_RAW / "h2s_rtma_full_matched_60_events.csv"

# Planned output file
rtma_matched_output_path = DATA_PROCESSED / "h2s_rtma_matched_events_clean.csv"

# Check files
print("H2S + RTMA audit setup")
print("=" * 60)

print("Current cleaned H2S file:")
print(current_h2s_path)
print("File exists:", current_h2s_path.exists())

print("\nRTMA matched input file:")
print(rtma_matched_input_path)
print("File exists:", rtma_matched_input_path.exists())

print("\nPlanned cleaned RTMA output:")
print(rtma_matched_output_path)

# Load files only if both exist
if current_h2s_path.exists() and rtma_matched_input_path.exists():
    h2s_current = pd.read_csv(current_h2s_path)
    h2s_rtma_raw = pd.read_csv(rtma_matched_input_path)

    print("\nCurrent cleaned H2S shape:", h2s_current.shape)
    print("RTMA matched raw shape:", h2s_rtma_raw.shape)

    print("\nCurrent H2S columns:")
    for col in h2s_current.columns:
        print("-", col)

    print("\nRTMA matched columns:")
    for col in h2s_rtma_raw.columns:
        print("-", col)

    print("\nCurrent H2S preview:")
    display(h2s_current.head())

    print("\nRTMA matched preview:")
    display(h2s_rtma_raw.head())
else:
    print("\nOne or both files are missing. Check the file names or folder location before moving forward.")
```

```python
h2s_current_check = h2s_current.copy()
h2s_rtma_check = h2s_rtma_raw.copy()

# Standardize datetime fields
h2s_current_check["datetime_local"] = pd.to_datetime(h2s_current_check["datetime_local"], errors="coerce")
h2s_rtma_check["datetime_local"] = pd.to_datetime(h2s_rtma_check["datetime_local"], errors="coerce")
h2s_rtma_check["datetime_utc"] = pd.to_datetime(h2s_rtma_check["datetime_utc"], errors="coerce")

# Standardize location and H2S values
h2s_current_check["location_clean"] = h2s_current_check["location_clean"].astype(str).str.strip().str.upper()
h2s_rtma_check["location_clean"] = h2s_rtma_check["location_clean"].astype(str).str.strip().str.upper()

h2s_current_check["h2s_ppb"] = pd.to_numeric(h2s_current_check["h2s_ppb"], errors="coerce")
h2s_rtma_check["h2s_ppb"] = pd.to_numeric(h2s_rtma_check["h2s_ppb"], errors="coerce")

# Create rounded H2S value for safer matching
h2s_current_check["h2s_ppb_round"] = h2s_current_check["h2s_ppb"].round(3)
h2s_rtma_check["h2s_ppb_round"] = h2s_rtma_check["h2s_ppb"].round(3)

# Recreate threshold flags for audit
h2s_current_check["valid_h2s_record"] = h2s_current_check["h2s_ppb"].notna()
h2s_current_check["elevated_ge_5"] = h2s_current_check["valid_h2s_record"] & (h2s_current_check["h2s_ppb"] >= 5)
h2s_current_check["elevated_ge_10"] = h2s_current_check["valid_h2s_record"] & (h2s_current_check["h2s_ppb"] >= 10)
h2s_current_check["orange_ge_30"] = h2s_current_check["valid_h2s_record"] & (h2s_current_check["h2s_ppb"] >= 30)

h2s_rtma_check["elevated_ge_5"] = h2s_rtma_check["h2s_ppb"].notna() & (h2s_rtma_check["h2s_ppb"] >= 5)
h2s_rtma_check["elevated_ge_10"] = h2s_rtma_check["h2s_ppb"].notna() & (h2s_rtma_check["h2s_ppb"] >= 10)
h2s_rtma_check["orange_ge_30"] = h2s_rtma_check["h2s_ppb"].notna() & (h2s_rtma_check["h2s_ppb"] >= 30)

print("H2S + RTMA match audit")
print("=" * 60)

print("Current cleaned H2S threshold counts:")
print("Total rows:", len(h2s_current_check))
print("Valid numeric H2S records:", h2s_current_check["valid_h2s_record"].sum())
print("Events >= 5 ppb:", h2s_current_check["elevated_ge_5"].sum())
print("Events >= 10 ppb:", h2s_current_check["elevated_ge_10"].sum())
print("Events >= 30 ppb:", h2s_current_check["orange_ge_30"].sum())

print("\nRTMA matched threshold counts:")
print("Total RTMA rows:", len(h2s_rtma_check))
print("Events >= 5 ppb:", h2s_rtma_check["elevated_ge_5"].sum())
print("Events >= 10 ppb:", h2s_rtma_check["elevated_ge_10"].sum())
print("Events >= 30 ppb:", h2s_rtma_check["orange_ge_30"].sum())

print("\nRTMA date/time coverage:")
print("Start datetime_local:", h2s_rtma_check["datetime_local"].min())
print("End datetime_local:", h2s_rtma_check["datetime_local"].max())
print("Unique dates:", h2s_rtma_check["datetime_local"].dt.date.nunique())
print("Unique stations:", h2s_rtma_check["location_clean"].nunique())

# Exact match check: datetime + station + H2S value
match_keys = ["datetime_local", "location_clean", "h2s_ppb_round"]

rtma_exact_match = h2s_rtma_check.merge(
    h2s_current_check[match_keys].drop_duplicates(),
    on=match_keys,
    how="left",
    indicator=True
)

exact_matches = (rtma_exact_match["_merge"] == "both").sum()
exact_nonmatches = (rtma_exact_match["_merge"] == "left_only").sum()

print("\nExact match check using datetime_local + location_clean + h2s_ppb:")
print("Exact matches:", exact_matches)
print("Exact non-matches:", exact_nonmatches)

# Looser match check: datetime + station only
loose_keys = ["datetime_local", "location_clean"]

rtma_loose_match = h2s_rtma_check.merge(
    h2s_current_check[loose_keys].drop_duplicates(),
    on=loose_keys,
    how="left",
    indicator=True
)

loose_matches = (rtma_loose_match["_merge"] == "both").sum()
loose_nonmatches = (rtma_loose_match["_merge"] == "left_only").sum()

print("\nLoose match check using datetime_local + location_clean only:")
print("Loose matches:", loose_matches)
print("Loose non-matches:", loose_nonmatches)

print("\nRTMA rows by station:")
display(h2s_rtma_check["location_clean"].value_counts().reset_index(name="rtma_rows"))

print("\nRTMA H2S value summary:")
display(h2s_rtma_check["h2s_ppb"].describe())

print("\nCurrent cleaned H2S events >= 10 ppb by station:")
display(
    h2s_current_check[h2s_current_check["elevated_ge_10"]]
    .groupby("location_clean")
    .agg(
        events_ge_10=("h2s_ppb", "size"),
        max_h2s_ppb=("h2s_ppb", "max")
    )
    .reset_index()
    .sort_values("events_ge_10", ascending=False)
)

print("\nRTMA exact non-matching rows, if any:")
display(
    rtma_exact_match[rtma_exact_match["_merge"] == "left_only"]
    [["datetime_local", "location_clean", "h2s_ppb", "h2s_level"]]
    .head(20)
)
```

Note to self:
The H2S + RTMA weather dataset is a weather-matched subset of the stronger elevated H2S events, defined as H2S >= 10 ppb. It is used to examine wind and weather conditions during higher-intensity H2S events, not all elevated H2S records.

```python
# Prepares the H2S + RTMA matched subset, where all events are H2S >= 10 ppb and have
# hourly weather/wind variables.

h2s_rtma_clean = h2s_rtma_raw.copy()

# Parse datetime fields
h2s_rtma_clean["datetime_local"] = pd.to_datetime(h2s_rtma_clean["datetime_local"], errors="coerce")
h2s_rtma_clean["datetime_utc"] = pd.to_datetime(h2s_rtma_clean["datetime_utc"], errors="coerce")
h2s_rtma_clean["date"] = pd.to_datetime(h2s_rtma_clean["date"], errors="coerce").dt.date

# Standardize location field
h2s_rtma_clean["location_clean"] = h2s_rtma_clean["location_clean"].astype(str).str.strip().str.upper()

# Convert key numeric columns
numeric_cols = [
    "target_lat",
    "target_lon",
    "h2s_ppb",
    "temperature_2m",
    "dewpoint_2m",
    "wind_u_10m",
    "wind_v_10m",
    "wind_direction_10m",
    "wind_speed_10m",
    "wind_gust_10m",
    "visibility_surface",
    "wind_speed_10m_mph",
    "wind_gust_10m_mph",
    "visibility_surface_km",
    "visibility_surface_miles",
    "temperature_2m_f",
    "dewpoint_2m_f"
]

for col in numeric_cols:
    if col in h2s_rtma_clean.columns:
        h2s_rtma_clean[col] = pd.to_numeric(h2s_rtma_clean[col], errors="coerce")

# Create clean event flags
h2s_rtma_clean["valid_h2s_record"] = h2s_rtma_clean["h2s_ppb"].notna()
h2s_rtma_clean["elevated_h2s_event"] = h2s_rtma_clean["valid_h2s_record"] & (h2s_rtma_clean["h2s_ppb"] >= 5)
h2s_rtma_clean["h2s_ge_10_event"] = h2s_rtma_clean["valid_h2s_record"] & (h2s_rtma_clean["h2s_ppb"] >= 10)
h2s_rtma_clean["orange_h2s_event"] = h2s_rtma_clean["valid_h2s_record"] & (h2s_rtma_clean["h2s_ppb"] >= 30)

# Add cleaned level labels based on H2S value
h2s_rtma_clean["h2s_level_clean"] = np.select(
    [
        h2s_rtma_clean["h2s_ppb"].isna(),
        h2s_rtma_clean["h2s_ppb"] < 5,
        (h2s_rtma_clean["h2s_ppb"] >= 5) & (h2s_rtma_clean["h2s_ppb"] < 30),
        h2s_rtma_clean["h2s_ppb"] >= 30
    ],
    [
        "Missing/Not visible",
        "Green (0-4.9 ppB)",
        "Yellow (5-29.9 ppB)",
        "Orange (30+ ppB)"
    ],
    default="Check"
)

# Add useful time fields
h2s_rtma_clean["year"] = h2s_rtma_clean["datetime_local"].dt.year
h2s_rtma_clean["month"] = h2s_rtma_clean["datetime_local"].dt.to_period("M").astype(str)
h2s_rtma_clean["hour"] = h2s_rtma_clean["datetime_local"].dt.hour

# Document the subset definition directly in the dataset
h2s_rtma_clean["rtma_match_subset_definition"] = "H2S events >= 10 ppb matched to hourly RTMA weather/wind variables"

# Organize most useful columns first
front_cols = [
    "datetime_local",
    "datetime_utc",
    "date",
    "year",
    "month",
    "hour",
    "location_clean",
    "target_lat",
    "target_lon",
    "h2s_ppb",
    "h2s_level_clean",
    "valid_h2s_record",
    "elevated_h2s_event",
    "h2s_ge_10_event",
    "orange_h2s_event",
    "temperature_2m_f",
    "dewpoint_2m_f",
    "wind_speed_10m_mph",
    "wind_direction_10m",
    "wind_gust_10m_mph",
    "visibility_surface_miles",
    "temperature_2m",
    "dewpoint_2m",
    "wind_u_10m",
    "wind_v_10m",
    "wind_speed_10m",
    "wind_gust_10m",
    "visibility_surface",
    "nighttime_flag",
    "analysis_period",
    "rtma_match_subset_definition"
]

# Keep remaining original columns after the organized front columns
remaining_cols = [col for col in h2s_rtma_clean.columns if col not in front_cols]
h2s_rtma_clean = h2s_rtma_clean[front_cols + remaining_cols].sort_values(
    ["datetime_local", "location_clean"]
).reset_index(drop=True)

print("Cleaned H2S + RTMA matched dataset created")
print("=" * 60)
print("Rows:", h2s_rtma_clean.shape[0])
print("Columns:", h2s_rtma_clean.shape[1])

print("\nDate/time coverage:")
print("Start datetime_local:", h2s_rtma_clean["datetime_local"].min())
print("End datetime_local:", h2s_rtma_clean["datetime_local"].max())
print("Unique dates:", h2s_rtma_clean["date"].nunique())
print("Unique stations:", h2s_rtma_clean["location_clean"].nunique())

print("\nEvent counts:")
print("Valid H2S records:", h2s_rtma_clean["valid_h2s_record"].sum())
print("H2S events >= 5 ppb:", h2s_rtma_clean["elevated_h2s_event"].sum())
print("H2S events >= 10 ppb:", h2s_rtma_clean["h2s_ge_10_event"].sum())
print("Orange H2S events >= 30 ppb:", h2s_rtma_clean["orange_h2s_event"].sum())

print("\nMissing values in key weather/wind columns:")
key_weather_cols = [
    "temperature_2m_f",
    "dewpoint_2m_f",
    "wind_speed_10m_mph",
    "wind_direction_10m",
    "wind_gust_10m_mph",
    "visibility_surface_miles"
]
print(h2s_rtma_clean[key_weather_cols].isna().sum())

print("\nPreview:")
display(h2s_rtma_clean.head())
```

```python
# now lets match the data point to stations and variables
print("H2S + RTMA cleaned QA summary")
print("=" * 60)

print("Overall dataset:")
print("Rows:", len(h2s_rtma_clean))
print("Unique dates:", h2s_rtma_clean["date"].nunique())
print("Unique stations:", h2s_rtma_clean["location_clean"].nunique())
print("Duplicate rows:", h2s_rtma_clean.duplicated().sum())

print("\nDuplicate checks using datetime + station:")
print(
    h2s_rtma_clean.duplicated(
        subset=["datetime_local", "location_clean"]
    ).sum()
)

print("\nStation-level summary:")
h2s_rtma_station_summary = (
    h2s_rtma_clean
    .groupby("location_clean", dropna=False)
    .agg(
        matched_events=("h2s_ppb", "size"),
        mean_h2s_ppb=("h2s_ppb", "mean"),
        median_h2s_ppb=("h2s_ppb", "median"),
        max_h2s_ppb=("h2s_ppb", "max"),
        orange_h2s_events=("orange_h2s_event", "sum"),
        mean_wind_speed_mph=("wind_speed_10m_mph", "mean"),
        median_wind_speed_mph=("wind_speed_10m_mph", "median"),
        max_wind_gust_mph=("wind_gust_10m_mph", "max"),
        mean_temperature_f=("temperature_2m_f", "mean"),
        mean_dewpoint_f=("dewpoint_2m_f", "mean"),
        mean_visibility_miles=("visibility_surface_miles", "mean")
    )
    .reset_index()
    .sort_values("matched_events", ascending=False)
)

display(h2s_rtma_station_summary)

print("\nWeather/wind variable summary:")
weather_summary_cols = [
    "h2s_ppb",
    "temperature_2m_f",
    "dewpoint_2m_f",
    "wind_speed_10m_mph",
    "wind_direction_10m",
    "wind_gust_10m_mph",
    "visibility_surface_miles"
]

display(h2s_rtma_clean[weather_summary_cols].describe())

print("\nH2S level counts:")
print(h2s_rtma_clean["h2s_level_clean"].value_counts(dropna=False))

print("\nEvents by month:")
display(
    h2s_rtma_clean
    .groupby("month")
    .agg(
        matched_events=("h2s_ppb", "size"),
        orange_h2s_events=("orange_h2s_event", "sum"),
        mean_h2s_ppb=("h2s_ppb", "mean"),
        max_h2s_ppb=("h2s_ppb", "max")
    )
    .reset_index()
)
```

```python
# lets exports the cleaned H2S + RTMA matched event subset and station/month summaries
# for the final project

h2s_rtma_overall_summary = pd.DataFrame({
    "metric": [
        "total_matched_events",
        "unique_dates",
        "unique_stations",
        "start_datetime_local",
        "end_datetime_local",
        "mean_h2s_ppb",
        "median_h2s_ppb",
        "max_h2s_ppb",
        "h2s_events_ge_10ppb",
        "orange_h2s_events_ge_30ppb",
        "mean_wind_speed_mph",
        "median_wind_speed_mph",
        "max_wind_gust_mph",
        "mean_temperature_f",
        "mean_dewpoint_f",
        "mean_visibility_miles"
    ],
    "value": [
        len(h2s_rtma_clean),
        h2s_rtma_clean["date"].nunique(),
        h2s_rtma_clean["location_clean"].nunique(),
        str(h2s_rtma_clean["datetime_local"].min()),
        str(h2s_rtma_clean["datetime_local"].max()),
        round(h2s_rtma_clean["h2s_ppb"].mean(), 3),
        round(h2s_rtma_clean["h2s_ppb"].median(), 3),
        round(h2s_rtma_clean["h2s_ppb"].max(), 3),
        int(h2s_rtma_clean["h2s_ge_10_event"].sum()),
        int(h2s_rtma_clean["orange_h2s_event"].sum()),
        round(h2s_rtma_clean["wind_speed_10m_mph"].mean(), 3),
        round(h2s_rtma_clean["wind_speed_10m_mph"].median(), 3),
        round(h2s_rtma_clean["wind_gust_10m_mph"].max(), 3),
        round(h2s_rtma_clean["temperature_2m_f"].mean(), 3),
        round(h2s_rtma_clean["dewpoint_2m_f"].mean(), 3),
        round(h2s_rtma_clean["visibility_surface_miles"].mean(), 3)
    ]
})

h2s_rtma_station_summary = (
    h2s_rtma_clean
    .groupby("location_clean", dropna=False)
    .agg(
        matched_events=("h2s_ppb", "size"),
        unique_dates=("date", "nunique"),
        mean_h2s_ppb=("h2s_ppb", "mean"),
        median_h2s_ppb=("h2s_ppb", "median"),
        max_h2s_ppb=("h2s_ppb", "max"),
        orange_h2s_events=("orange_h2s_event", "sum"),
        mean_wind_speed_mph=("wind_speed_10m_mph", "mean"),
        median_wind_speed_mph=("wind_speed_10m_mph", "median"),
        max_wind_gust_mph=("wind_gust_10m_mph", "max"),
        mean_temperature_f=("temperature_2m_f", "mean"),
        mean_dewpoint_f=("dewpoint_2m_f", "mean"),
        mean_visibility_miles=("visibility_surface_miles", "mean")
    )
    .reset_index()
    .sort_values("matched_events", ascending=False)
)

h2s_rtma_month_summary = (
    h2s_rtma_clean
    .groupby("month", dropna=False)
    .agg(
        matched_events=("h2s_ppb", "size"),
        unique_stations=("location_clean", "nunique"),
        mean_h2s_ppb=("h2s_ppb", "mean"),
        median_h2s_ppb=("h2s_ppb", "median"),
        max_h2s_ppb=("h2s_ppb", "max"),
        orange_h2s_events=("orange_h2s_event", "sum"),
        mean_wind_speed_mph=("wind_speed_10m_mph", "mean"),
        max_wind_gust_mph=("wind_gust_10m_mph", "max")
    )
    .reset_index()
)

# Output paths
h2s_rtma_overall_summary_path = TABLES / "h2s_rtma_overall_summary.csv"
h2s_rtma_station_summary_path = TABLES / "h2s_rtma_station_summary.csv"
h2s_rtma_month_summary_path = TABLES / "h2s_rtma_month_summary.csv"

# Save files
h2s_rtma_clean.to_csv(rtma_matched_output_path, index=False)
h2s_rtma_overall_summary.to_csv(h2s_rtma_overall_summary_path, index=False)
h2s_rtma_station_summary.to_csv(h2s_rtma_station_summary_path, index=False)
h2s_rtma_month_summary.to_csv(h2s_rtma_month_summary_path, index=False)

# Reload cleaned file for QA
h2s_rtma_reloaded = pd.read_csv(rtma_matched_output_path)

print("Saved H2S + RTMA cleaned files")
print("=" * 60)

print("\nProcessed data file:")
print(rtma_matched_output_path)
print("File exists:", rtma_matched_output_path.exists())
print("Reloaded shape:", h2s_rtma_reloaded.shape)

print("\nSummary tables:")
for path in [
    h2s_rtma_overall_summary_path,
    h2s_rtma_station_summary_path,
    h2s_rtma_month_summary_path
]:
    print(path)
    print("File exists:", path.exists())

print("\nReloaded QA:")
print("Rows:", len(h2s_rtma_reloaded))
print("Unique dates:", h2s_rtma_reloaded["date"].nunique())
print("Unique stations:", h2s_rtma_reloaded["location_clean"].nunique())
print("H2S events >= 10 ppb:", h2s_rtma_reloaded["h2s_ge_10_event"].sum())
print("Orange H2S events >= 30 ppb:", h2s_rtma_reloaded["orange_h2s_event"].sum())

print("\nStation summary:")
display(h2s_rtma_station_summary)
```
